# Linear Regression
Predicting a continuos value

#### Equation : y = Q1*x + Q0
Q1 is slope
Q0 is intercept

#### Psedocode :
1) Initialize Q=[Q1 Q0]
2) Measure how good our Q is -- Reduced Error in training data
3) Update Q (using gradient descent algo) so that J(Q) i.e Mean Squared Error decreases        

### Gradient Descent (in general) :
Used to find local minima in any given function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2 
print(X,Y)

#Plotting the function
plt.plot(X,Y)           
plt.xlabel('X') 
plt.ylabel('F(x)')
plt.show()
```

```python
# Given a function of f(x) = (x-5)^2, we want to value of x that minimizes f using Gradient Descent.

x=0
lr=0.1 
error=[]

# 50 steps in downward direction
plt.plot(X,Y)

for i in range(50):
    grad=2*(x-5) 
    x= x - lr*grad             #Update x in the downward direction of gradient
    y = (x-5)**2               #Calculate error value
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
#Plotting the error values
plt.plot(error)
```

## Gradient descent for Linear Regression

$Q = [Q_0  Q_1]$

$J(Q) = \frac{1}{m} \sum_{i=1}^m (y_{-}^{(i)} - y^{(i)})^2$

Hypothesis Function:

$y_{-} = Q_0 + Q_1x$

After Derivation parameters :

$Q_0 = Q_0 - \alpha \frac{1}{m} \sum_{i=1}^m (y_{-}^{(i)} - y^{(i)})$

$Q_1 = Q_1 - \alpha \frac{1}{m} \sum_{i=1}^m (y_{-}^{(i)} - y^{(i)})x^{(i)}$

Where:
- $m$ is the number of training examples
- $\alpha$ is the learning rate
- $x^{(i)}$ is the i-th input feature
- $y^{(i)}$ is the i-th target value
- $y_{-}^{(i)}$ is the predicted output
