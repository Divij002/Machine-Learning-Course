# K Nearest Neigbours impementation
- Simple ML supervised algorithm used for classifcation and Regression
- No training is used for this algo (O(1) time used)
- The majority in k nearest points decides which class the point belongs to
- there are no parameters
- This algo is used as a Baseline

### Implementation :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification

# Generate 399 samples, 2 features, 2 classes
X, Y = make_classification(
    n_samples=399,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=7
)

print("X shape:", X.shape)
print("Y shape:", Y.shape)

# Show first 5 rows
print("\nFirst 5 X rows:\n", X[:5])
print("\nFirst 5 Y labels:\n", Y[:5])
```

```python
# Visualize the dataset
plt.scatter(X[:,0], X[:,1], c=Y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.show()
```

```python
query_x = np.array([-0.5,-0.7])
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(query_x[0],query_x[1],color='red')
plt.show()
```

#### Algorithm

```python
# Using Eucledean Distance
def dist(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5):
    vals = []
    m = X.shape[0]

    for i in range(m): 
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))

    vals = sorted(vals)
    vals = vals[:k]
    
    vals = np.array(vals)
    
    #print max val
    new_vals = np.unique(vals[:,1],return_counts=True)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    return pred
```

```python
query_x = np.array([-1,2])
knn(X,Y,query_x)
```
