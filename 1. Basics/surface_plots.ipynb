# Surface Plots | Data Visualisation

surface plots are used to
- Visualise Loss Functions in ML and DL
- Visualise State or State Value Functions in Reinforcement Learning

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#a = np.array([1,2,3]) , b = np.array([4,5,6,7])

a = np.arange(-1,1,0.02)
b=a

a,b = np.meshgrid(a,b)
```

```python
fig = plt.figure()
axes = fig.add_subplot(111, projection='3d')
axes.plot_surface(a,b, a**2 + b**2, cmap='viridis')
plt.show()
```

# Contour Plot

```python
fig = plt.figure()
axes = fig.add_subplot(111, projection='3d')
axes.contour(a,b, a**2 + b**2, cmap='viridis')
plt.show()
```
