# Linear Regression using skLearn

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
X,y = make_regression(n_samples=1000, n_features=2, noise=10, random_state=1, n_informative=1)

print(X.shape, y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0], y, color='blue', marker='o', s=10)

plt.subplot(1,2,2)
plt.scatter(X[:,1], y, color='red', marker='o', s=10)
plt.show()
```

```python
# Output plane y = Q0 + Q1*x1 + Q2*x2

from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')

ax.scatter3D(X[:,0], X[:,1], y, color='green', marker='o', s=10)
plt.title("3D Scatter plot")
plt.show()
```

```python
# Impementation of Linear Regression using sklearn

model = LinearRegression()
model.fit(X,y)
```

```python
print("Q1 and Q2 :", model.coef_ ,"\n Q0 : ", model.intercept_)
```

```python
model.predict([X[0], X[1]])
```

```python
y[0], y[1]
```

```python
model.score(X,y)  # R^2 value
```
