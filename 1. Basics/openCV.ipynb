# OpenCV

```python
import cv2
import matplotlib.pyplot as plt
```

Each image is a matrix which is 3 levels deep.
For example 100px * 100px * 3 --- these 3 layers are of RGB colors which have their own matrix with values

```python
# Read an image from a file
img = cv2.imread('images/monk.jpg')
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert BGR to Grayscale

print(img.shape)                          #shape of the image - it will show (height, width, channels)
```

```python
plt.imshow(rgb_image)                         #show the image
#plt.axis('off')                               #turn off the axis
```

```python
#Cropping image
crop_image = rgb_image[100:1900, 100:1900]      # Crop the image (y1:y2, x1:x2)
plt.imshow(crop_image)                          #show the cropped image
```

Image pixels should be less, The more the image pixels the higher space it takes in the matrix

```python
#Resizing the image
crop_image = cv2.resize(crop_image, (100, 100))                  # Resize the image to 1000x1000 pixels
plt.imshow(crop_image)  
plt.title('Monk'+str(crop_image.shape))  
plt.show()                      
```
