# Basics of OpenCV

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read an Image

img = cv2.imread('./dog_img.png')  # Reads BGR as default

# Convert BGR to RGB
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display both images
plt.imshow(img)
plt.show()
plt.imshow(rgb_img)
plt.show()
```

```python
# Image is a matrix of (pixel, pixel, RGB)
print(img.shape)
print(img)
```
